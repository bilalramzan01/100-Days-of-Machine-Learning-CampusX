# How to create virtaul or new environment

- open anaconda -> environment -> U can see all the installed libraries there.
- At start the base environment opened.
- It consist of 100s od pre installed libraries, so when we deploy the project we have to deploy all these. thats why when we working on project we create new virtual environment.

 ## Create environment

 1. open anaconda -> environment -> Create new environment
 2. using CMD: Anaconda prompt -> conda create --name CampusX -> conda activate campusx
    - So first of all we install jupter notebook as its a also library.
      conda install -c jupyter notebook
    - install required libraries
 3. Delete: By click on remove or by cmd conda remove --name campusx
