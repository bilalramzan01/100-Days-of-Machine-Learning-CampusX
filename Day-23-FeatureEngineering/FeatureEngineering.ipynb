## Feature Engineering

Feature Engineering is the process of using domain knowledge to extract and transform features from raw data to improve ML model performance.

**STEPS:**

1. Datasets → Data Retrieval – raw data is collected/pulled from its source
2. Data Preparation (the core stage) — includes three parts:
3. Data Processing & Wrangling – cleaning, handling missing values, fixing formats
4. Feature Extraction & Engineering – creating new meaningful features from raw data
5. Feature Scaling & Selection – normalizing/standardizing values and picking the most relevant features
6. Modeling – the prepared data is fed into a Machine Learning Algorithm to build a model
7. Model Evaluation & Tuning – model performance is checked and hyperparameters are adjusted
8. Deployment & Monitoring – the final model is deployed and tracked in production
9. Re-iteration loop – if performance isn't satisfactory, the process loops back to Data Preparation and repeats

## Feature Engineering branches into 4 main categories:

1. Feature Transformation — modifying existing features to make them more suitable for ML models. Includes:

- Missing Value Imputation – filling in missing data (Days 35–40 you covered)
- Handling Categorical Features – encoding text categories into numbers (ordinal/one-hot encoding)
- Outlier Detection – identifying and treating extreme values (Z-score, IQR methods)
- Feature Scaling – standardizing/normalizing value ranges (StandardScaler, etc.)

2. Feature Construction — creating new features from existing ones (e.g., combining "height" and "weight" into "BMI")

3. Feature Selection — choosing the most relevant features and dropping irrelevant/redundant ones (like we did with the correlation analysis on your cricket dataset — dropping "Year")

4. Feature Extraction — deriving new, more compact features from raw or high-dimensional data (e.g., PCA, extracting features from text/images)
