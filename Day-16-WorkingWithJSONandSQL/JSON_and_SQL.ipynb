# Working with SQL

**SQL (Structured Query Language)** is the standard language used to communicate with relational databases — used to create, read, update, and delete data stored in structured tables (rows and columns).

**Core idea**

Data in SQL databases is organized into tables (like spreadsheets), where each table has columns (fields) and rows (records). SQL is how you query, filter, join, and manipulate this data.
