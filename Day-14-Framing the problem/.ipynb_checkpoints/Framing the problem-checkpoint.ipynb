# Framing the Problem

How to Frame a Machine Learning Problem | How to plan a Data Science Project Effectively

**Business Problem** 
 Netflix -> you are at data science position -> In a meeting you have to suggest how we can increase the revenue.

1. What if we reduce the churn rate (Churn rate means how many customers leaving the platform)
2. We can increate the customers

churn rate is 4% in netflix right now.

Now I have to decrease the churn rate.

**Goal**: churn rate should be reduced from 4% to 3.7%.



# Problem steps

1. Identify the customers who are leaving the platform
   - Imediate plan ( Offer the discount for next month subscription)
   - finding the point why he is leaving
2. Supervised learning, we have to identify of each customers of is leaving and who is not.
   - Identify the % who much the person is going to leave and you have to offer discount accordingly
   - Now the problem is move to regression problem as we are finding %.
3. I have model who predict the next month churn rate
   - we will take help from this model.
4. Data collection:
   - Check watch time, Search time, and browsing time
   - content left in the middle
5. Metrics to measures:
   - Prediction is going right or not
   - prediction of % is similar or not
6. Batch or online learning:
   - Decide which leaning is to use
   - I will prefare online learning because data is continously coming.
7. Check your assumptions
   - Confirm the factors I assume
   - will all the factors or models work on all the places or locations
