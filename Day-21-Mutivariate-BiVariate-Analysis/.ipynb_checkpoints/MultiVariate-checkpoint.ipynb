# 1. Scatter plot (Numerical - Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'], hue=tips['sex'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'], hue=tips['sex'],style=tips['smoker'])
```

## 2.Bar plot (Numerical - Categorical)

```python
titanic.head(5)
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

## 3.Boxplot

```python
sns.boxplot(x=titanic['Pclass'],y=titanic['Age'], hue=titanic['Survived'])
```

## Distplot

```python
sns.distplot(titanic['Survived']==0, hist=False)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

## Heat Map (Categorical - Categorical)

```python
titanic.head(5)
```

```python
pd.crosstab(titanic['Pclass'], titanic ['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic ['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
```
