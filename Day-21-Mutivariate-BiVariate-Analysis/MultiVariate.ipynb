```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

## 2.Bar plot (Numerical - Categorical)

```python
titanic.head(5)
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

## 3.Boxplot

```python
sns.boxplot(x=titanic['Pclass'],y=titanic['Age'], hue=titanic['Survived'])
```

## Distplot

```python
sns.distplot(titanic['Survived']==0, hist=False)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

## Heat Map (Categorical - Categorical)

```python
titanic.head(5)
```

```python
pd.crosstab(titanic['Pclass'], titanic ['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic ['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
```

```python
# Female saved more then males

titanic.groupby('Sex')['Survived'].mean() * 100
```

```python
titanic.groupby('Embarked')['Survived'].mean() * 100
```

## Cluster map

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
