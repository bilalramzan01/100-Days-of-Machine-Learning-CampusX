# Fetching Data From an API (Application programming interface)
An API (Application Programming Interface) is exactly this — it's a messenger between you (your Python code) and someone else's server/database. You ask for data, the API fetches it and hands it back to you, without you needing to know how their internal system works.
- API used to communicate two software
- API is nothing but data pipelines

```python
import requests

urls = {
    "JSON": "https://jsonplaceholder.typicode.com/posts/1",
    "CSV": "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv",
    "XML": "https://www.w3schools.com/xml/note.xml",
    "HTML": "https://example.com",
}

for label, url in urls.items():
    response = requests.get(url)
    content_type = response.headers.get('Content-Type', 'Unknown')
    print(f"{label}: Content-Type = {content_type}")
```
