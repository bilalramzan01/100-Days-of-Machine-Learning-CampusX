```python
web = soup.find_all("div", class_="maso-item")
```

```python
name = []

for i in web:
    course_name = i.find('a').find('img')['alt']
    name.append(course_name)

print(name)
```

```python
len(name)
```

```python
df = pd.DataFrame({'course_name': name})
df.tail()
```

```python
links = []

for i in web:
    course_link = i.find('a')['href']
    links.append(course_link)

# print(links)
```

```python
df = pd.DataFrame({"Course_Links": links})
```
